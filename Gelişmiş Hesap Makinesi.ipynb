```python
import math

# --- MODÜLER FONKSİYONLAR (İş Mantığı) ---

def toplama(veriler):
    return sum(veriler)

def cikarma(veriler):
    # İlk sayıdan diğer tüm sayıları sırayla çıkarır
    sonuc = veriler[0]
    for sayi in veriler[1:]:
        sonuc -= sayi
    return sonuc

def carpma(veriler):
    # Tüm sayıları birbiriyle çarpar
    sonuc = 1
    for sayi in veriler:
        sonuc *= sayi
    return sonuc

def bolme(veriler):
    # İlk sayıyı diğerlerine sırayla böler
    sonuc = veriler[0]
    for sayi in veriler[1:]:
        if sayi == 0:
            return " Hata: Sıfıra bölme yapılamaz!"
        sonuc /= sayi
    return round(sonuc, 2)

def ortalama_hesapla(veriler):
    return sum(veriler) / len(veriler)

def kare_analizi(veriler):
    return [round(n**2, 2) for n in veriler]

def karekok_analizi(veriler):
    return [round(math.sqrt(abs(n)), 2) for n in veriler]

def raporla(islem_adi, sonuc):
    print(f"\n[ANALİZ SONUCU] {islem_adi}: {sonuc}")

# --- ANA PROGRAM DÖNGÜSÜ ---
def main():
    print("="*45)
    print(" PROFESYONEL ANALİTİK İŞLEM MOTORU")
    print("="*45)

    while True:
        try:
            girdi = input("\nSayıları boşlukla girin (veya çıkış için 'q'): ")
            
            if girdi.lower() == 'q':
                print("Sistem kapatılıyor... Analizler tamamlandı.")
                break

            sayilar = [float(x) for x in girdi.split()]
            
            if not sayilar:
                print(" Veri girişi algılanamadı.")
                continue

            print("\nKomutlar: [t: Topla] [-: Çıkar] [*: Çarp] [/: Böl] [o: Ortalama] [k: Kareler] [s: Karekök]")
            komut = input("Seçiminiz: ").lower().strip()

            # İşlem Katmanı
            if komut == "t":
                raporla("Toplam Değer", toplama(sayilar))
            elif komut == "-":
                raporla("Çıkarma Sonucu (S1-S2-...)", cikarma(sayilar))
            elif komut == "*":
                raporla("Çarpım Sonucu", carpma(sayilar))
            elif komut == "/":
                raporla("Bölme Sonucu", bolme(sayilar))
            elif komut == "o":
                raporla("Aritmetik Ortalama", ortalama_hesapla(sayilar))
            elif komut == "k":
                raporla("Sayıların Kareleri", kare_analizi(sayilar))
            elif komut == "s":
                raporla("Karekök Analizi (Mutlak Değer)", karekok_analizi(sayilar))
            else:
                print("❌ Tanımlanamayan komut.")

        except ValueError:
            print(": Geçersiz veri formatı! Lütfen sadece sayı girin.")

if __name__ == "__main__":
    main()
```
